```python
import pandas as pd
import numpy as np
from kafka import KafkaProducer
import json
import random


# Bước 1: Đọc file CSV
def read_csv(file_path):
    data = pd.read_csv(file_path, sep='|')
    return data


# Bước 2: Thay đổi ngẫu nhiên giá trị của các cột "INTENT_NAME", "INTENT_CONFIDENCE", "NLU_THRESHOLD"
def modify_columns_randomly(data):
    # Chọn ngẫu nhiên 10% của tập dữ liệu để thay đổi
    n_rows_to_modify = int(0.1 * len(data))
    random_indices = np.random.choice(data.index, n_rows_to_modify, replace=False)

    random_intent_names = ['Intent_A', 'Intent_B', 'Intent_C']

    nlu_threshold_value = round(np.random.uniform(0.7, 1.0), 1)

    for idx in random_indices:
        data.at[idx, 'INTENT_NAME'] = random.choice(random_intent_names)

        data.at[idx, 'INTENT_CONFIDENCE'] = round(np.random.uniform(0.0, 1.0), 4)  # Giá trị ngẫu nhiên từ 0.0 đến 1.0

        data.at[idx, 'NLU_THRESHOLD'] = nlu_threshold_value

    return data


# Bước 3: Gửi dữ liệu lên Kafka
def send_to_kafka(topic, data, kafka_broker):
    producer = KafkaProducer(
        bootstrap_servers=kafka_broker,
        value_serializer=lambda v: json.dumps(v).encode('utf-8')
    )

    # Biến đổi từng dòng của data thành json và gửi lên Kafka
    for index, row in data.iterrows():
        producer.send(topic, row.to_dict())

    producer.flush()



```

```python
file_path = "data/bot_ddvc_hcm_bot_predict.csv"

data = read_csv(file_path)

# Thay đổi ngẫu nhiên giá trị của các cột "INTENT_NAME", "INTENT_CONFIDENCE", "NLU_THRESHOLD"
modified_data = modify_columns_randomly(data)

display(modified_data)
```

```python
# Thông tin Kafka
kafka_broker = '192.168.1.53:9092'
kafka_topic = 'bot_ddvc_hcm_bot_predict'

# Gửi dữ liệu đã thay đổi lên Kafka
send_to_kafka(kafka_topic, modified_data, kafka_broker)

print("Dữ liệu đã được gửi lên Kafka!")
```
